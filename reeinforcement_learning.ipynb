```python
import numpy as np
import random
```

```python
n_arms = 3
#we have 3 slot machines
```

```python
true_rewards = [0.2,0.5,0.8]
```

```python
success = np.ones(n_arms)
failure = np.ones(n_arms)
```

```python
episodes = 1000
total_reward = 0
```

```python
for episodes in range(episodes):
    #machines id guessing the best arm to pull
    sampled_values = [
        #generating random probability on the basis of old success and failure rate
        np.random.beta(success[i], failure[i])
        for i in range(n_arms)
    ]
    #pick the highes sampled value
    action = np.argmax(sampled_values)
    #if random probability is less than true values then increase the reward
    reward = 1 if random.random() < true_rewards[action] else 0

    if reward == 1:
        success[action] += 1
    else:
        failure[action] += 1
```

```python
print("successes:", success)
print("failures:", failure)
print("Total Reward:", total_reward)
print("Best Arm Learned:", np.argmax(success))
```
