DBSCAN - Destiny based spatial clustering of applications with noise

- outliers(noise) part of data but are not included in clusters 
- to make the data more efficient

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Mall_Customers .csv')
df
```

```python
X = df.iloc[:,[3,4]].values
```

```python
X = StandardScaler().fit_transform(X)
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5 )
clusters = dbscan.fit_predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=clusters, cmap= 'viridis', s=50)
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('DBSCAN Clustering')
plt.show()
```

```python
#finding outliers
outliers = X[clusters == -1] # -1 because DBSCAN labels outliers as -1
print(f'Number of outliers detected: {len(outliers)}')
```

```python
plt.scatter(X[:,0], X[:,1], c=clusters, cmap='viridis', s=50)
plt.scatter(outliers[:,0], outliers[:,1], c='red', s=100, label='Outliers', edgecolors='k')
plt.title('DBSCAN Clustering with Outliers Highlighted')
```
